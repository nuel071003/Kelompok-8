**Mengisi dan Mempersiapkan untuk Pelatihan Model**

```python
# Mengisi semua kolom numerik dengan rata-rata masing-masing kolom
imputer_mean = SimpleImputer(strategy='mean')
foot[foot.select_dtypes(include=['float64', 'int64']).columns] = imputer_mean.fit_transform(foot.select_dtypes(include=['float64', 'int64']))
```

```python
# Mempersiapkan Fitur dan Target untuk Model
x = foot.drop(columns=['overall_rating','name','full_name','positions','nationality','birth_date','preferred_foot','body_type','national_team_position','national_team'])
y = foot['overall_rating']
```

**Melatih (Training) Pada Data Tersebut**

```python
# Membagi Data Pelatihan dan Pengujian Model
from sklearn.model_selection import train_test_split
train_x, test_x , train_y, test_y = train_test_split(x, y, test_size = 0.10, random_state=40)
```

```python
# Memilih Kolom Numerik dari Dataframe
num = foot.select_dtypes('number').reset_index(drop = True)
```

```python
# Inisialisasi Random Forest
clf = RandomForestClassifier(n_estimators=3000, random_state=500)
```

```python
# Melatih Model Random Forest dengan Data
new_var = clf.fit(train_x, train_y)
new_var
```

**Melakukan sebuah Pengujian Data**

```python
# Hasil Prediksi Random Forest
predictions = clf.predict(test_x)
```

```python
# Pengujian Data dan Akurasi Data
accuracy = accuracy_score(test_y, predictions)
print(f"Accuracy: {accuracy:2f}")
```
